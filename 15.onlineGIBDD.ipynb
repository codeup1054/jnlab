```python
# 2022-03-31 
import json
import requests

tm()

url =  'https://api.onlinegibdd.ru/v3/partner_fines/'

lst = dfr['auto_id']['id'] # [:500]

n = 200 

cars_id_chunks = [lst[i:i + n] for i in range(0, len(lst), n)]

chunk_num = int(len(lst)/n)

dft = pd.DataFrame()

for cars_id in  cars_id_chunks:
    autos_ids = ",".join(cars_id)
    
    chunk_num = chunk_num - 1

    payload = {
    "status": "all",
    'autos_ids': autos_ids,
    'per_page': 10,    
    'date_bill_start': "2021-08-01"
    }
    
    headers = {'content-type': 'application/json', 'Accept-Charset': 'UTF-8', 'Authorization': 'Bearer ec67734d-eb00-4da7-bd47-61b95518cd25' }
    r = requests.post(url, json=payload, headers=headers)
    fines_json = r.json()
    
    auto_list = fines_json['data']['auto_list']
    
    for o in auto_list:
        offense_list = auto_list[o]['offense_list']
        auto_id = auto_list[o]['auto_id']
        auto_number =dfr['auto_id'].loc[dfr['auto_id']['id'] == auto_id, 'auto_number'].iloc[0]
        auto_region = dfr['auto_id'].loc[dfr['auto_id']['id'] == auto_id, 'auto_region'].iloc[0]


        for k, oll in offense_list.items():
#             print (len(oll))
#             disp (pd.DataFrame.from_dict(oll, orient='index').T)
            dfn = pd.DataFrame.from_dict(oll, orient='index').T
            dfn['auto_id'] = auto_id
            dfn['auto_number'] =auto_number
            dfn['auto_region'] =auto_region
             
            dft = pd.concat([dft,dfn])
    
    
    tm([chunk_num, dft.shape, len(fines_json['data']['auto_list'])])


tm([">>>", len(fines_json['data']['auto_list'])])
dfr['offence'] = dft
disp(dft[:3])
dft.count()
```
